```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

```python
# determines what the user currently has in their hand
cards = ['8S','JH','3S','3D','QD','8S','8D']
def determine_hand(cards):
  # royal flush
  flush_conditions = [['10C', 'JC', 'QC', 'KC', 'AC'], ['10D', 'JD', 'QD', 'KD', 'AD'], ['10H', 'JH', 'QH', 'KH', 'AH'], ['10S', 'JS', 'QS', 'KS', 'AS']]
  if flush_conditions in cards:
    return "royal_flush"

  #flush check
  c_count, s_count, d_count, h_count = 0, 0, 0, 0
  for card in cards:
    if card.count('S') == 1:
      s_count += 1
    elif card.count('C') == 1:
      c_count += 1
    elif card.count('D') == 1:
      d_count += 1
    elif card.count('H') == 1:
      h_count

  # straight / straight flush
  straight_cards = []
  for card in cards:
    if "10" in card:
      straight_cards.append(10)
    elif "J" in card:
      straight_cards.append(11)
    elif "Q" in card:
      straight_cards.append(12)
    elif "K" in card:
      straight_cards.append(13)
    elif "A" in card:
      straight_cards.append(14)
      straight_cards.append(1)
    else:
      straight_cards.append(int(card[0]))
  straight_cards = sorted(straight_cards)


  length = len(straight_cards)
  cards_in_straight = []
  straight = False
  for i in range(0, length-4):
    if straight_cards[i]+4 == straight_cards[i+1]+3 == straight_cards[i+2]+2 == straight_cards[i+3]+1 == straight_cards[i+4]:
      if c_count >= 5 or h_count >= 5 or d_count >= 5 or s_count >= 5:
        return "straight flush"
      else:
        straight = True
        break

  #check for four of a kind
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 4:
      return "four of a kind"

  #full house
  fh_pt1 = False
  fh_pt2 = False
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 3:
      fh_pt1 = True
    elif straight_cards.count(straight_card) == 2:
      fh_pt2 = True
    if fh_pt1 and fh_pt2:
      return "full house"

  #flush
  if c_count >= 5 or h_count >= 5 or d_count >= 5 or s_count >= 5:
    return "flush"

  #straight
  if straight:
    return "straight"

  #three of a kind
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 3:
      return "three of a kind"

  #two pair
  pair1 = False
  num_pair = -1
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 2:
      if pair1 and straight_card != num_pair:
        return "two pair"
      else:
        num_pair = straight_card
        pair1 = True

  #pair
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 2:
      return "pair"

  #high card
  cards.sort()
  return "high card"
```

```python
# solves for the odds of them getting a different hand
def find_odds(cards):
  odds = {}
  l = len(cards)
  #straight stuff
  straight_cards = []
  for card in cards:
    if "10" in card:
      straight_cards.append(10)
    elif "J" in card:
      straight_cards.append(11)
    elif "Q" in card:
      straight_cards.append(12)
    elif "K" in card:
      straight_cards.append(13)
    elif "A" in card:
      straight_cards.append(14)
      straight_cards.append(1)
    else:
      straight_cards.append(int(card[0]))
  straight_cards = sorted(straight_cards)

  #flush stuff
  c_count, s_count, d_count, h_count = 0, 0, 0, 0
  for card in cards:
    if card.count('S') == 1:
      s_count += 1
    elif card.count('C') == 1:
      c_count += 1
    elif card.count('D') == 1:
      d_count += 1
    elif card.count('H') == 1:
      h_count
  # pair
  pair = False
  for card in straight_cards:
    if straight_cards.count(card) == 2:
      odds["pair"] = "100%"
      pair = True
  if not pair:
    if l == 5:
      odds["pair"] = "58.6%"
    elif l == 6:
      odds["pair"] = "39.1%"

  #two pair
  pair1 = False
  num_pair = -1
  two_pair = False
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 2:
      if pair1 and straight_card != num_pair:
        two_pair = True
        odds["two pair"] = "100%"
        break
      else:
        num_pair = straight_card
        pair1 = True
  if not two_pair:
    if pair1:
      if l == 5:
        odds["two pair"] = "40.2%"
      if l == 6:
        odds["two pair"] = "26.1%"
    else:
      if l == 5:
        odds["two pair"] = "8.3%"
      if l == 6:
        odds["two pair"] = "0%"

  #3 of a kind
  threepair = False
  twopair = False
  num_two_pair = 0
  spec_card = -1
  for card in straight_cards:
    if straight_cards.count(card) == 2:
      twopair = True
      if spec_card != card:
        num_two_pair += 1
      spec_card = card
    if straight_cards.count(card) == 3:
      odds["three of a kind"] = "100%"
      threepair = True
  if not threepair:
    if twopair:
      if l == 5:
        odds["three of a kind"] = "8.5%"
      elif l == 6 and num_two_pair == 2:
        odds["three of a kind"] = "8.7%"
      elif l == 6:
        odds["three of a kind"] = "4.3%"
    else:
      if l == 5:
        odds["three of a kind"] = "1.4%"
      if l == 6:
        odds["three of a kind"] = "0%"

  #flush
  if c_count >= 5 or h_count >= 5 or d_count >= 5 or s_count >= 5:
    odds["flush"] = "100%"
  else:
    if c_count < 3 and h_count < 3 and d_count < 3 and s_count < 3:
      odds["flush"] = "0%"
    else:
      if l == 5:
          if c_count == 4 or h_count == 4 or d_count == 4 or s_count == 4:
            odds["flush"] = "36.4%"
          else:
            odds["flush"] = "4.2%"
      elif l == 6:
        if c_count < 4 and h_count < 4 and d_count < 4 and s_count < 4:
          odds["flush"] = "0%"
        else:
          odds["flush"] = "19.5%"

  #full house
  fh_pt1 = False
  fh_pt2 = False
  full_house = False
  num_pair = 0
  special_card = -1
  for straight_card in straight_cards:
    if straight_cards.count(straight_card) == 3:
      fh_pt1 = True
    elif straight_cards.count(straight_card) == 2:
      fh_pt2 = True
      if special_card != straight_card:
        num_pair += 1
      special_card = straight_card
    if fh_pt1 and fh_pt2:
      full_house = True
      odds["full house"] = "100%"
  if not full_house:
    if num_pair == 2:
      if l == 5:
        odds["full house"] = "24.1%"
      elif l == 6:
        odds["full house"] = "13%"
    elif fh_pt2:
      if l == 5:
        if num_pair == 2:
          odds["full house"] = "24.1%"
        else:
          odds["full house"] = "1.6%"
      elif l == 6:
        if num_pair == 2:
          odds["full house"] = "8.7%"
        else:
          odds["full house"] = "0%"
    else:
      odds["full house"] = "0%"

  #4 of a kind
  max_cards = 0
  max_hit = 0
  for straight_card in straight_cards:
    num = straight_cards.count(straight_card)
    if num == 4:
      odds["four of a kind"] = "100%"
    if num > max_cards:
      max_cards = num
      max_hit = 1
    elif num == max_cards:
      max_hit += 1
  if l == 5:
    if max_cards < 2:
      odds["four of a kind"] = "0%"
    elif max_cards == 2 and max_hit == 2:
      odds["four of a kind"] = "0.1%"
    elif max_cards == 2 and max_hit == 1:
      odds["four of a kind"] = "0.05%"
    elif max_cards == 3:
      odds["four of a kind"] = "4.2%"
  elif l == 6:
    if max_cards < 3:
      odds["four of a kind"] = "0%"
    elif max_cards == 3 and max_hit == 2:
      odds["four of a kind"] = "4.3%"
    elif max_cards == 3 and max_hit == 1:
      odds["four of a kind"] = "2.1%"

  # straight
  max_streak = -1
  current_streak = 0
  for i in range(0, len(straight_cards)-1):
    if straight_cards[i]+1 == straight_cards[i+1]:
      current_streak += 1
    else:
      if current_streak > max_streak:
        max_streak = current_streak
      current_streak = 0

  if max_streak >= 5:
    odds["straight"] = "100%"
  elif l == 5:
    if max_streak < 3:
      odds["straight"] = "0%"
    elif max_streak == 3:
      odds["straight"] = "2.9%"
    elif max_streak == 4:
      odds["straight"] = "31.4%"
  elif l == 6:
    if max_streak < 4:
      odds["straight"] = "0%"
    elif max_streak == 4:
      odds["straight"] = "17.3%"

  #straight flush
  if max_streak >= 5 and (c_count >= 5 or h_count >= 5 or d_count >= 5 or s_count >= 5):
    odds["straight flush"] = "100%"
  elif l == 5:
    if max_streak < 3:
      odds["straight flush"] = "0%"
    elif max_streak == 3 and c_count >= 3 or h_count >= 3 or d_count >= 3 or s_count >= 3:
      odds["straight flush"] = "0.1%"
    elif max_streak == 4 and c_count >= 4 or h_count >= 4 or d_count >= 4 or s_count >= 4:
      odds["straight flush"] = "7.3%"
  elif l == 6:
    if max_streak < 4:
      odds["straight flush"] = "0%"
    elif max_streak == 4 and c_count >= 4 or h_count >= 4 or d_count >= 4 or s_count >= 4:
      odds["straight flush"] = "4.3%"

  #royal flush
  num_spade, num_club, num_heart, num_diamond = 0,0,0,0
  req_spade = ["KS", "QS", "JS", "AS", "10S"]
  req_club = ["KC", "QC", "JC", "AC", "10C"]
  req_heart = ["KH", "QH", "JH", "AH", "10H"]
  req_diamond = ["KD", "QD", "JD", "AD", "10D"]
  for card in cards:
    if card in req_spade:
      num_spade += 1
    elif card in req_club:
      num_club += 1
    elif card in req_heart:
      num_heart += 1
    elif card in req_diamond:
      num_diamond += 1

  if num_spade == 5 or num_club == 5 or num_heart == 5 or num_diamond == 5:
    odds["royal flush"] = "100%"
  elif l == 5:
    if num_spade < 3 or num_club < 3 or num_heart < 3 or num_diamond < 3:
      odds["royal flush"] = "0%"
    elif num_spade == 3 or num_club == 3 or num_heart == 3 or num_diamond == 3:
      odds["royal flush"] = "0.1%"
    elif num_spade == 4 or num_club == 4 or num_heart == 4 or num_diamond == 4:
      odds["royal flush"] = "4.2%"
  elif l == 6:
    if num_spade < 4 or num_club < 4 or num_heart < 4 or num_diamond < 4:
      odds["royal flush"] = "0%"
    elif num_spade == 4 or num_club == 4 or num_heart == 4 or num_diamond == 4:
      odds["royal flush"] = "2.1%"

  return odds
```
